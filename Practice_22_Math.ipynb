# PYTHON MATH:

Python math module provides built-in mathematical functions and constants for performing advanced mathematical calculations.

---
| Function  | Meaning         |
| --------- | --------------- |
| `sqrt()`  | square root     |
| `pow()`   | power           |
| `floor()` | round down      |
| `ceil()`  | round up        |
| `fabs()`  | remove minus    |
| `pi`      | circle constant |
| `e`       | growth constant |


```python
# Ex-1 square root - math.sqrt():

import math
print(math.sqrt(144))
```

```python
# Ex-2 Power – math.pow():
import math
print(math.pow(3,3))
```

```python
# Ex-3 Floor – math.floor():
import math
print(math.floor(4.8))

# Floor = below

# 4.8 → 4
```

```python
# Ex-4 Ceil – math.ceil():
import math
print(math.ceil(4.1))

# Ceiling = top

# 4.1 → 5
```

```python
# Ex-5 Absolute Value – math.fabs():
import math
print(math.fabs(-24))
      
# Special float-only need → math.fabs()
```

```python
# Ex-6: abs():
x = -12
print(abs(x))

# Default → abs()
```

```python
# Ex-7 Value of π (Pi) – math.pi:
import math

print(math.pi)
```

```python
# Ex - 8 Value of e – math.e: Mathematical constant
import math

print(math.e)
```

### REAL-LIFE PRACTICE:

```python
# Example 1: Finding square root of land area
import math
area = 144
root_area = math.sqrt(area)
print(root_area)
```

```python
# Example 2: Power calculation for compound growth
import math
base = 2
power = 5
result = math.pow(base, power)
print(result)
```

```python
# Example 3: Rounding down bill units
import math
units = 12.9
final_units = math.floor(units)
print(final_units)
```

```python
# Example 4: Rounding up pages required
import math
pages = 7.1
required_pages = math.ceil(pages)
print(required_pages)
```

```python
# Example 5: Distance calculation using absolute value
distance = -25
actual_distance = abs(distance)
print(actual_distance)
```
