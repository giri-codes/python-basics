# PYTHON DATA TYPES:

```python
# Numeric data types (int, float, complex):

age = 25
salary = 100000.50
power = 3 + 5j

print(type(age))
print(type(salary))
print(type(power))
```

```python
# String data type:
name = "Giri"
city = "Bangalore"

print(type(name))
print(type(city))
```

```python
# List data type:
# Ex - 1
bikes = ["jawa", "royal enfield", "hero", "honda"]
print(type(bikes))

# Ex - 2
numbers = [10, 20, 30, 40]

numbers.append(50)
print(numbers)
print(type(numbers))
```

```python
# Tuple data type:
#Ex - 1
colors = ("red", "green", "blue")
print(type(colors))

#Ex - 2

single = (10,)
print(type(single))
```

```python
# Set data type:
#Ex - 1

numbers = {10, 20, 30, 30, 40}
print(type(numbers))

#Ex - 2

fruits = {"apple", "banana", "mango"}
fruits.add("orange")
print(fruits)
print(type(fruits))
```

```python
#Dictionary data type:
student = {
    "name": "Giri",
    "age": 25,
    "course": "Python"
}

print(student)
print(type(student))
```

```python
#Boolean data types:
#Ex - 1
is_on = True
is_off = False

print(type(is_on))
print(type(is_off))

#Ex - 2
x = 10
y = 20

print(x > y)
print(x == y)
```

```python
#None data type:
g = None
print(g)
print(type(g))
```
