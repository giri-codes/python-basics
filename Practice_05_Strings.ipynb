# PYTHON STRINGS:

```python
# Strings basics:
name = "Giri"
print(name)

message = "Hii Python"
print(message)

sentence = "Python is easy to learn"
print(sentence)
```

```python
a = "I am a developer"
print(a[0:4])
```

```python
for b in "Gireesh":
    print(b)
```

```python
c = "ML Engineer"
print(len(c))
```

```python
# Multiline strings:
text = """Python is easy
Python is powerful
Python is popular"""
print(text)
```

```python
# triple single quotes:
message = '''Hello Giri,
Welcome to Python learning.
Practice daily.'''
print(message)
```

```python
# Multiline string with formatting:
name = "Giri"

info = f"""
Name   : {name}
Course : Python
Level  : Beginner
"""
print(info)
```

```python
# Check strings:
#using in
text = "Python developer"

print("Python" in text)     # True
print("Java" in text)       # False
```

```python
# Using not in:
t1 = "Java developer"
print("Java" not in t1)  
print("Python" not in t1)
```

```python
t2 = "Python developer"

print("python" in t2)     # False
print("Python" in t2)    # True
```

```python
# String slicing:
g ="Python Programming"

print(g[0])        # First character
print(g[0:6])      # Python
print(g[7:18])     # Programming
print(g[:6])       # Python
print(g[7:])       # Programming
print(g[-11:])     # Programming
print(g[::-1])     # Reverse string
```

```python
# Modify strings:
f = " Giri, The ML Engineer "

print(f.upper())
print(f.lower())
print(f.strip())
print(f.replace("Giri", "Gireesh"))
print(f.split())
```

```python
# String concatenation:
first = "Python"
second = "Developer"

result = first + " " + second
print(result)
```

```python
# Wrong:
# age = 25
# print("Age is " + age)  ‚ùå

# Correct:
age = 25
print("Age is", age)
```

```python
# Format strings:
#using f strings:
name = "Giri"
age = 25

print(f"My name is {name} and my age is {age}")
```

```python
# Using format():
name = "Gireesh"
age = 27

print("My name is {} and my age is {}".format(name, age))
```

```python
# Using % operator:

name = "Giri B"
age = 26

print("My name is %s and my age is %d" % (name, age))
```

```python
# Escape characters:

print("Hello\nPython")      # New line
print("Hello\tPython")      # Tab space
print("I\'m learning Python")
print("He said \"Python is easy\"")
print("C:\\Users\\Giri")
```
