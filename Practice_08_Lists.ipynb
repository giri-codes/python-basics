## PYTHON LOOP LISTS:

```python
# for loop:
i1 =["a1", "b1", "c1", "d1"]

for a in i1:
    print(a)
```

```python
# loop through index numbers:

j1 = ["Giri", "developer", "ML Engineer", 100000]

for i in range(len(j1)):
    print(j1[i])
```

```python
# while loop:
k1 = ["Giri", "developer", "ML Engineer", 100000]

i = 0
while i < len(k1):
    print(k1[i])
    i = i + 1
```

```python
# looping using list comprehension:
l1 = [2,4,6,8,10]

[print(a) for a in l1]
```

## PYTHON LIST COMPREHENSION:

```python
seeds = ["chana", "moong", "horse gram", "green peas", "chia", "fenugreek"]

seeds1 = [x for x in seeds if "e" in x]

print(seeds1)
```

```python
seeds = ["chana", "moong", "horse gram", "green peas", "chia", "fenugreek"]

seeds1 = [x for x in seeds if x != "moong"]

print(seeds1)
```

```python
# Wit no if statement:
seeds = ["chana", "moong", "horse gram", "green peas", "chia", "fenugreek"]

seeds1 = [x for x in seeds]

print(seeds1)
```

```python
list1 = [x for x in range(12)]
print(list1)
```

```python
list2 = tuple(x for x in range(12))
print(list2)
```

```python
list3 = {x for x in range(12)}
print(list3)
```

```python
list4 = [x for x in range(10) if x < 9]
print(list4)
```

```python
# upper() case:
lentils = ["chana", "moong", "horse gram", "green peas", "chia", "fenugreek"]

lentils1 = [x.upper() for x in lentils]

print(lentils1)
```

```python
# setting all values to "Peanuts":

lentils = ["chana", "moong", "horse gram", "green peas", "chia", "fenugreek"]

lentils1 = ["Peanuts" for x in lentils]

print(lentils1)
```

```python
# Return "TVS" insted of "honda":

bikes = ["hero", "jawa", "royal", "hero", "honda"]

new_bikes = [x if x != "honda" else "TVS" for x in bikes]

print(new_bikes)
```

## PYTHON SORT LISTS:

```python
#Ascending:
num1 = [6,23,8,5,2,3,5]

num1.sort()
print(num1)
```

```python
# Descending:
num2 = [12,54,36,89,98,32,12,51]

num2.sort(reverse = True) # reverse=True → big → small
print(num2)
```

```python
#sort string list:
bikes = ["hero", "jawa", "royal", "honda"]
bikes.sort()
print(bikes)
```

```python
#sort with mixed case letters:

names = ["Giri", "apple", "Banana", "cat"]
names.sort()
print(names)
```

```python
# Case-insensitive sort:
names = ["Giri", "apple", "Banana", "cat"]
names.sort(key=str.lower) # str.lower → ignore case
print(names)
```

```python
# Sort without changing original list:
nums = [3, 1, 4, 2]

new_nums = sorted(nums)

print(nums)
print(new_nums)

# sort() → changes original

# sorted() → creates new list
```

```python
# Sort list of strings by length:
words = ["python", "hi", "developer", "go"]
words.sort(key=len)
print(words)
```

## PYTHON COPY LISTS:

```python
# USE copy() method()
c = ["I", "am", "a", "developer"]

d = c.copy()
print(d)
```

```python
# By using list() method:
c = ["I", "am", "a", "developer"]

d = list(c)
print(d)
```

```python
# By slice operator:
a = [10, 100, 1000, 50, 500, 5000]
b = a[:]

print(b)
```

```python
a = [10, 20, 30]
b = a.copy()

b.append(40)

print(a)
print(b)
```

```python
a = ["hero", "honda", "tvs"]
b = a[:]

b.remove("honda")

print(a)
print(b)
```

```python
a = [1, 2, 3, 4]
b = list(a)

b[0] = 100

print(a)
print(b)
```

```python
# copy inside a list comprehension:
a = [2, 4, 6]
b = [x for x in a]

b.append(8)

print(a)
print(b)
```

## PYTHON JOIN LISTS:

```python
# By using + operator:
m1 = [2,4,6,8]
m2 = ["a", "c", "e", "g"]

m3 = m1 + m2
print(m3)
```

```python
# By append method:
m1 = [2,4,6,8]
m2 = ["a", "c", "e", "g"]

for x in m2:
    m1.append(x)

print(m1)
```

```python
# By extend() method:
m1 = [2,4,6,8]
m2 = ["a", "c", "e", "g"]

m1.extend(m2)
print(m1)
```

```python
team_a = ["Ravi", "Anita"]
team_b = ["Kiran", "Megha"]

full_team = team_a + team_b
print(full_team)
```

```python
morning_tasks = ["yoga", "coding"]
evening_tasks = ["gym", "reading"]

morning_tasks.extend(evening_tasks)
print(morning_tasks)
```

```python
course_modules = ["Python", "SQL"]
extra_modules = ["Power BI", "Tableau"]

course_modules.append(extra_modules)
print(course_modules)
```
