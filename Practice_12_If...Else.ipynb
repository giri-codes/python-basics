# PYTHON IF ELSE:

### IF:

`if condition:`

     statement


```python
# Ex -1:
temp = 24
if temp > 23:
    print("Fever detected")
```

```python
# Ex - 2:
salary = 25000

if salary >= 20000:
    print("Eligible for loan")
```

```python
# Ex - 3:
steps = 10000

if steps >= 8000:
    print("Daily goal completed")
```

### IF ELSE:

`if condition:`

    statement
    
`else:`

    statement


```python
# Ex - 1:
marks = 55

if marks >= 40:
    print("Pass")
else:
    print("Fail")
```

```python
#Ex - 2:
age = 16

if age >= 18:
    print("Can vote")
else:
    print("Cannot vote")
```

```python
#Ex - 3:
balance = 1500

if balance >= 2000:
    print("Payment successful")
else:
    print("Insufficient balance")
```

### ELIF:

`if condition1:`

    statement
    
`elif condition2:`

    statement
    
`else:`

    statement


```python
#X - 1:
marks = 72

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

```python
#Ex - 2:
time = 14

if time < 12:
    print("Good morning")
elif time < 18:
    print("Good afternoon")
else:
    print("Good evening")
```

```python
#Ex - 3:
speed = 85

if speed > 100:
    print("Overspeeding")
elif speed > 60:
    print("Normal speed")
else:
    print("Slow driving")
```

### Logical Operators `(and, or, not)`:

```python
# and:
age = 25
has_id = True

if age >= 18 and has_id:
    print("you can enter")
```

```python
# or:
day = "Sunday"

if day == "Saturday" or day == "Sunday":
    print("Weekend")
```

```python
# not:
is_logged_in = False

if not is_logged_in:
    print("Please login first")
```

### Shorthand if:

```python
#Ex - 1: if
g = 27
r = 22

if g > r: print("g is greater")
```

```python
#Ex - 2 : if...else:
marks = 34
print("pass") if marks >= 40 else print("fail")
```

```python
#Ex - 3:
number = 8
print("even") if number % 2 == 0 else print("odd")
```

### Nested if:

**Syntax**

`if condition1:`

    if condition2:
    
        statement


```python
# Ex -1 :
age = 27
has_id = True

if age >= 18:
    if has_id:
        print("Entry allowed")
    else:
        print("Id required")
```

```python
#Ex - 2:
user_id = "giri"
password = "1234"

if user_id == "giri":
    if password == "1234":
        print("login successful")
    else:
        print("wrong password")

else:
    print("invalid user")
```

```python
#Ex - 3:
balance = 10000000

if balance > 0:
    if balance >= 50000:
        print("Minimum balance maintained")
    else:
        print("High balance warning")
```
