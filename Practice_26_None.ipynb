# PYTHON NONE:

```python
# Example 1: Variable not initialized yet
# Real-life usage: Data not loaded yet
result = None

if result is None:
    print("Result not ready")
```

```python
# Example 2: Function returns nothing
# Real-life usage: Action-only functions
def greet():
    print("Hello")

output = greet()
print(output)   # None
```

```python
# Example 3: Default function argument
# Real-life usage: Optional parameters
def send_email(message=None):
    if message is None:
        print("No message provided")
    else:
        print("Sending:", message)

send_email()
send_email("Hi")
```

```python
# Example 4: Checking None correctly
# Real-life usage: Safe condition checks
value = None

if value is None:
    print("Value is None")
```

```python
# Example 5: Wrong way vs right way
# Real-life usage: Interview trap
x = None

print(x == None)    # works, but NOT recommended
print(x is None)    # correct way
```

```python
# Example 6: None in if condition (truthy / falsy)
# Real-life usage: Control flow
data = None

if data:
    print("Has data")
else:
    print("No data")   # This runs
```

```python
# Example 7: Distinguishing None from 0
# Real-life usage: Calculations vs missing values
x = None
y = 0

if x is None:
    print("x has no value")

if y == 0:
    print("y is zero")
```

```python
# Example 8: Using None as placeholder
# Real-life usage: Login systems
user = None

if user is None:
    print("User not logged in")
```

```python
# Example 9: Resetting a variable
# Real-life usage: Clearing cache / session
score = 100
score = None

print(score)
```

```python
# Example 10: Function returning None intentionally
# Real-life usage: Validation functions
def validate(age):
    if age < 0:
        return None
    return age

print(validate(-5))
print(validate(25))
```

```python
# Example 11: None with lists
# Real-life usage: Marking missing data
values = [10, None, 30]

for v in values:
    if v is None:
        print("Missing value")
    else:
        print(v)
```

```python
# Example 12: None + try/except together
# Real-life usage: Safe data processing
try:
    result = int("abc")
except ValueError:
    result = None

if result is None:
    print("Conversion failed safely")
```
