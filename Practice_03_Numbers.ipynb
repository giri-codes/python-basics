# PYTHON NUMBERS:

`int`, `float`, `complex`

```python
#int()
#Ex - 1

age = 25
year = 2026

print(age)
print(year)
```

```python
#Ex - 2

a = 15
b = 4

print(a + b)
print(a - b)
print(a * b)
print(a // b)
```

```python
#Ex - 3
x = -10
y = abs(x)

print(y)
```

```python
#float()
#Ex - 1
price = 99.75
discount = 12.5

print(price - discount)
```

```python
#Ex - 2
distance = 7.5
time = 1.5

speed = distance / time
print(speed)
```

```python
#ex - 3
value = 10.0
print(type(value))
```

```python
# Integer + Float Together:
#Ex - 1
a = 10
b = 2.5

print(a + b)
print(type(a + b))
```

```python
#Ex - 2
marks = 450
subjects = 6

average = marks / subjects
print(average)
```

```python
#Type conversion:
#Ex - 1
x = 10
y = float(x)

print(y)
print(type(y))
```

```python
#Ex - 2
price = "100"
price = int(price)

print(price + 50)
```

```python
#Complex numbers:
#Ex - 1
a = 3 + 4j
print(a)
```

```python
#Ex - 2
a = 2 + 3j
b = 1 + 2j

print(a + b)
```

```python
#Ex - 3
num = 5 + 6j

print(num.real)
print(num.imag)
```
