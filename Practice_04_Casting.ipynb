# PYTHON CASTING:

```python
# Casting to integer:
# Ex - 1
x = 10.7
y = int(x)

print(y)
```

```python
# Ex - 2
price = "250"
price = int(price)

print(price + 50)
```

```python
# Ex - 3
value = -8.9
print(int(value))
```

```python
# Casting to float:
# Ex - 1
a = 10
b = float(a)

print(b)
print(type(b))
```

```python
# Ex - 2
marks = "87.5"
marks = float(marks)

print(marks + 2.5)
```

```python
# Ex - 3
num = 5
print(num / 2)
```

```python
# Casting to String:
# Ex - 1
age = 25
msg = str(age)

print("Age is " + msg)
```

```python
# Ex - 2
salary = 100000
print("Salary:", str(salary))
```

```python
# Ex - 3
price = 99.5
print("Price is " + str(price))
```

```python
# Casting to complex:
# Ex - 1
x = 5
y = complex(x)

print(y)
print(type(y))
```

```python
# Ex - 2
a = 3
b = 4

z = complex(a, b)
print(z)
```

```python
# Mixed casting:
# Ex - 1
a = "10"
b = 5

result = int(a) + b
print(result)
print(type(result))
```

```python
# Ex - 2
x = 7.5
y = "2"

print(x * float(y))
```

```python
# Ex - 3
value = 100
value = str(value)

print(type(value))
```

### Invalid casting:
**print(int("10.5"))  # ValueError**

`Why Error?`

`int() expects whole numbers only`

`"10.5" contains a decimal point`

`Decimal numbers belong to float, not int`



**print(int("Giri"))  # ValueError**

`Why Error?`

`"Giri" contains letters`

`int() needs numeric digits only`

`Letters cannot be converted to a number`
