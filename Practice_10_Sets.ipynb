# PYTHON SETS:

`Access` → loop / in

`Add` → add, update

`Remove` → remove, discard, pop

`Delete` set → del

`Join` → union

`Common` → intersection

`Difference` → difference

`Different only` → symmetric_difference

`Locked` → frozenset

```python
the_set = {"giri", "java", "royal"}
print(the_set)
```

```python
# duplicates not allowed:
set1 = {"giri", "java", "royal", "giri"}
print(set1)
```

```python
#True and 1 considered as same value:
set2 = {"giri", "java", "royal", True, 1, 2}
print(set2)
```

```python
# False and 0 is considered as same values:
set3 = {"giri", "java", "royal", False, True, 0}
print(set3)
```

```python
# Length of a set:

set4 = {"giri", "java", "royal", False, True, 0}
print(len(set4))
```

```python
# Set data types:any data types can be allowed

set5 = {"lion", "tiger", "cheetah"}
set6 = {1, 5, 7, 9, 3}
set7 = {True, False, False}

print(set5)
print(set6)
print(set7)
```

```python
# Mixed datatypes:
set8 = {"Giri", 27, True, "Male", 42}
print(set8)
print(type(set8))
```

```python
# Set constructor:

set9 = set(("Giri", 27, True, "Male", 42))
print(set9)
```

### ACCESS SET ITEMS:

```python
# Using for loop:
set10 = {"giri", "java", "royal", True, 1, 2}

for x in set10:
    print(x)
```

```python
#Check item exists or not:

print("jawa" in set10)
print("giri" in set10)
```

```python
# Attendence check:
g = {"Ravi", "Radha", "Kiran"}
print("Ravi" in g)
```

```python
# Available colors:
colors = {"red", "blue", "green"}

for c in colors:
    print(c)
```

### ADD SET ITEMS:

```python
# Add new city:

cities = {"Bangalore", "Mysore"}
cities.add("Haveri")
print(cities)
```

```python
# Add new subjects:

subjects = {"Maths", "Science", "English"}
subjects.update({"English", "Computer"})
print(subjects)
```

```python
# Add product categories:
categories = {"Mobile", "Laptop"}
categories.add("Tablet")
categories.add("Earbuds")
print(categories)
```

### REMOVE SET ITEMS:

```python
# Remove absent student:
students = {"Ravi", "Anu", "Kiran"}
students.remove("Anu")
print(students)
```

```python
# Safe removal:
items = {"pen", "pencil"}
items.discard("eraser")
print(items)
```

```python
# Random removal:
numbers = {1, 2, 3}
numbers.pop()
print(numbers)
```

```python
# del keyword:
# delete temporary data:
temp_data = {101, 102}
del temp_data
```

```python
# Delete cancelled order:
order_items = {"shirt", "jeans"}
del order_items
```

```python
#Delete unused variable:
cache = {"a", "b"}
del cache
```

### LOOP SETS:

```python
# Loop numbers:

n = {1,12,3,14,5,16,7,18,9}
for a in n:
    print(a)
```

```python
# Loop employee IDs:

emp_ids = {101, 102, 103, 104, 105, 106, 107, 108, 109, 110}
for i in emp_ids:
    print(i)
```

```python
# Loop permisiions:
permissions = {"read", "write"}
for p in permissions:
    print(p)
```

### JOIN SETS:

```python
# Two class students:

class1 = {"A", "B"}
class2 = {"B", "C"}
print(class1.union(class2))
```

```python
# Two branch products:

store1 = {"milk", "bread"}
store2 = {"bread", "butter"}
print(store1 | store2)
```

```python
# Skills set:
skills1 = {"Python", "SQL"}
skills2 = {"SQL", "PowerBI"}
print(skills1.union(skills2))
```

```python
# Intersection:
# Common students:

a = {"Ravi", "Anu"}
b = {"Anu", "Kiran"}
print(a.intersection(b))
```

```python
# Common apps:
p1 = {"WhatsApp", "YouTube"}
p2 = {"YouTube", "Instagram"}
print(p1 & p2)
```

```python
# Common numbers:

n1 = {1,2,2,1,3}
n2 = {2,2,1,1,3,4,5}
print(n1.intersection(n2))
```

```python
# Intersection_update:
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set1.intersection_update(set2)

print(set1)
```

```python
# difference():
# Extra students:
a = {"A", "B", "C"}
b = {"B"}
print(a.difference(b))
```

```python
# Missing apps:
phone1 = {"WhatsApp", "YouTube"}
phone2 = {"WhatsApp"}
print(phone1 - phone2)
```

```python
# Skill gap:
required = {"Python", "SQL", "Excel"}
known = {"Python"}
print(required.difference(known))
```

```python
# Difference_update:
# Students who passed vs failed
passed = {"Ravi", "Anu", "Kiran"}
failed = {"Anu"}

passed.difference_update(failed)
print(passed)
print(failed)
```

```python
# Symmetric difference:
# Two classes:
a = {"A", "B", "C"}
b = {"C", "D"}
print(a.symmetric_difference(b))
```

```python
# Two phone apps:

p1 = {"ChatGPT", "YouTube", "Github"}
p2 = {"YouTube", "Telegram", "Github"}
print(p1 ^ p2)
```

```python
# Team players:

t1 = {"Rohit", "Virat"}
t2 = {"Virat", "Dhoni"}
print(t1.symmetric_difference(t2))
```

### FROZENSET:

```python
# Fixed rules:

rules = frozenset(["age", "id"])
print("age" in rules)
```

```python
# Account types:

types = frozenset(["savings", "current"])
for t in types:
    print(t)
```

```python
# Dictionary key usage:
key = frozenset([1, 2])
data = {key: "value"}
print(data)
```
